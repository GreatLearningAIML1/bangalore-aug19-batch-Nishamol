# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
      Goal_diff_count = GoalsFor - GoalsAgainst

```python
pd.options.display.max_rows = 100  #set to print all the teams with their Goal Difference
```

```python
laliga['GoalsFor'] = laliga['GoalsFor'].astype(int)
```

```python
laliga['GoalsAgainst'] = laliga['GoalsAgainst'].astype(int)
```

```python
# Below function finds the Goal Difference and prints all the team with goal difference and also prints highest/smallest Goal Difference.
#two ways of finding the max and min goal difference among which one approach is commented.
def Goal_diff_count(x):
    x['GoalDiff'] = x['GoalsFor'] - x['GoalsAgainst']
    print(x[['Team','GoalDiff']])
    print("Team that had maximum goal difference")
    #print(laliga[['Team','GoalDiff']].nlargest(1,'GoalDiff'))  #builtin function to get max goal difference. 
    print(x[x['GoalDiff'] == x['GoalDiff'].max()][['Team','GoalDiff']],"\n")   # this will check and get the row of the maximum Goal diff and prints Team with Goal Diff of that particular row.
    print("Team that had minimum goal difference ")
    #print(laliga[['Team','GoalDiff']].nsmallest(1,'GoalDiff'))
    print(x[x['GoalDiff'] == x['GoalDiff'].min()][['Team','GoalDiff']])
    

    
```

```python
Goal_diff_count(laliga)
```

```python
laliga['GoalDiff'] = laliga['GoalsFor'] - laliga['GoalsAgainst']
#laliga[['Team','GoalDiff']]
laliga[['Team','GoalDiff']].nlargest(1,'GoalDiff')
```

# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
laliga['GamesWon'] = laliga['GamesWon'].astype(int)
laliga['GamesPlayed'] = laliga['GamesPlayed'].astype(int)
```

```python
#calculating winning % as per the given formula.
laliga['Winning Percent'] = (laliga['GamesWon']/laliga['GamesPlayed'])*100   
```

```python
laliga['Winning Percent'].fillna(0.0,inplace= True)  #used for addressing numerical error.
```

```python
laliga.tail()  #Just to check if the numerical error is addressed.
```

```python
laliga[['Team','Winning Percent']].nlargest(5,'Winning Percent')  #Prints the top 5 teams with Winning Percent.
```
